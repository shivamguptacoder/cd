```python
Table1 ={
('S','me'):'NP VP',
('S','I'):'NP VP',
('S','you'):'NP VP',
('S','India'):'NP VP',
('S','Australia'):'NP VP',
('S','Steve'):'NP VP',
('S','John'):'NP VP',
('S','the'):'NP VP',
('S','a'):'NP VP',
('S','an'):'NP VP',

('NP','me'): 'P',
('NP','I'): 'P',
('NP','you'): 'P',
('NP','India'): 'PN',
('NP','Australia'): 'PN',
('NP','Steve'): 'PN',
('NP','John'): 'PN',
('NP','the'): 'D N',
('NP','a'): 'D N',
('NP','an'): 'D N',


('VP','is') :'V NP',
('VP','want') :'V NP',
('VP','won') :'V NP',
('VP','played') :'V NP',

('N','Championship'):'Championship',
('N','ball'):'ball',
('N','toss'):'toss',

('V','is'): 'is',
('V','want'): 'want',
('V','won'): 'won',
('V','played'): 'played',

('P','me'):'me',
('P','I'):'I',
('P','you'):'you',

('PN','India'):'India',
('PN',' Australia'):' Australia',
('PN', 'Steve'):'Steve',
('PN','John'):'John',


('D','the'):'the',
('D','a'):'a',
('D','an'):'an'
}

def parse(user_input,start_symbol,parsingTable):
    print(user_input)
    flag = 0
    
    lst = list(user_input.split(" "))
    lst.append('$')
    stack = []
    
    stack.append(start_symbol)

    input_len = len(user_input)
    buffer = 0

    while len(stack) > 0:
        print(stack)
        top = stack[len(stack)-1]
        print()
        print("Top =>",top)
        
        current_input = lst[buffer]
        
        print("Current_Input => ",current_input)

        if top == current_input:
            stack.pop()
            buffer = buffer + 1
            
        else:
            key = top , current_input
            print(key)
            if key not in parsingTable:
                flag = 1
                break

            value = parsingTable[key]
            print(value)
            if value !='#':
                
                value = list(value.split(" "))
                value = value[::-1]
                stack.pop()
                for element in value:
                    stack.append(element)
            else:
                stack.pop()

    if flag == 0:
        print("String accepted!")
    else:
        print("String not accepted!")

start = input("Enter the Start Symbol : ")
parse('India won the Championship',start,Table1)
```
